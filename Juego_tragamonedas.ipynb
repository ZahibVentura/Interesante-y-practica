##Explicación del Juego de Tragamonedas
El siguiente programa simula un pequeño juego de tragamonedas, también conocido como tragaperras. En este juego, el usuario ingresa la cantidad de dinero que desea depositar y elige un multiplicador para aumentar sus ganancias. Sin embargo, el costo por cada tirada aumenta proporcionalmente al multiplicador seleccionado.

###Reglas del Juego
Los posibles resultados que permiten al usuario ganar son los siguientes:


*   Par: Dos números del mismo tipo. Por ejemplo, (2, 2, 1) o (1, 3, 1)....
*   Tercia: Tres números del mismo tipo. Por ejemplo, (1, 1, 1).
*   Escalera: Tres números consecutivos. Por ejemplo, (1, 2, 3).

###Pagos y Costos
Los pagos de las ganancias para el multiplicador x1 (el multiplicador por defecto del programa) son los siguientes:


*   Par: $0.5

*   Tercia: $1

*   Escalera: $2


El costo por tirada para el multiplicador x1 es de $0.5. Si el jugador desea aumentar sus ganancias seleccionando un multiplicador mayor, el costo por tirada también aumentará proporcionalmente.

###Consideraciones Adicionales
Es importante tener en cuenta que, aunque el multiplicador aumenta las ganancias potenciales, también incrementa el riesgo, ya que el costo por tirada es mayor. Por lo tanto, el jugador debe gestionar su dinero de manera estratégica para maximizar sus posibilidades de ganar.




```python
import random as rd

#En esta parte del codigo se crea el input del dinero y el multiplicador
#junto a esto se consideran las condiciones por si el usuario ingresa datos no permitidos
try:
    dinero = float(input('Deposite su dinero: '))
    if dinero < 0:
        print("El dinero no puede ser negativo. Por favor, ingrese un número positivo.")
        exit()
    print('Elija su multiplicador')
    print('1. x1')
    print('2. x2')
    print('3. x4')
    print('4. x8')
    print('5. x16')
    Multi = int(input('Presiones el número que corresponda: '))
    if Multi < 1 or Multi > 5:
        print("Multiplicador no válido. Debe ser un número entre 1 y 5.")
        exit()
except ValueError:
    print("Entrada no válida. Por favor, ingrese un número.")
    exit()

#Establecemos un precio de juego o por jugada
costo_jugada = 2**(Multi-2)
print('Nota: El costo de cada jugada es: ',costo_jugada,)

#Creamos nuestra maquina de juego
while dinero > 0:
  enter = int(input('Presione 1 para jugar o presione 0 para salir: '))
  print('\n')
  if enter == 1:

    if dinero < costo_jugada:
        print("No tienes suficiente dinero para jugar.")
        break
    dinero -= costo_jugada
    print(f'Se descontó: {costo_jugada} al dinero actual, resta: {dinero}')
    X = []
    for i in range (3):
      x = rd.randint(1,3)
      X.append(x)
    #Se definen los pagos dependiendo los resultados
    if X[0] == X[1] and X[1] == X[2]:
      ganado = 1*Multi
      dinero += ganado
      print('Resulatdo: Tercia ')
      print('Estado: Gana $',ganado)
    elif X[0] == X[1] or X[0] == X[2] or X[1] == X[2]:
      ganado = 0.5*Multi
      dinero += ganado
      print('Resultado: Par ')
      print('Estado: Gana $',ganado)
    elif X[0] < X[1] and X[1] < X[2]:
      ganado = 2*Multi
      dinero += ganado
      print('Resultado: Escalera')
      print('Estado: Gana $',ganado)
    else:
      print('Resultado: Pierde')

    #Imprimimos y mostramos en pantalla que obtuvo el jugador
    print(f'| {X[0]} | {X[1]} | {X[2]} | \n')
    print('Dinero actual: ',dinero,'\n')

  elif enter ==0:
    break
#La ganacias finales
print('Dinero final: ',dinero)

```
