```python
import pandas as pd
import numpy as np
from datetime import date
#----------
#Cargamos el archivo excel
Estatus = "ESTATUS CLIENTES.xlsx"
datos = pd.read_excel(Estatus)
print(datos)
#-----
ESTATUS1 = "ESTATUS CTAS.xlsx"
datos2= pd.read_excel(ESTATUS1)
print(datos2)
#-------
ESTATUS2 = "OPERACIONES.xlsx"
datos3= pd.read_excel(ESTATUS2)
print(datos3)
#--------

#FECHA_DATOS---- 1)
def fecha(mes):
  try:
    if mes == 10:
      return date(2024,mes,31)
    elif mes == 11:
      return date(2024,mes,30)
    elif mes == 12:
      return date(2024,mes,31)
    else:
      return "Mes no valido"
  except TypeError:
    return "Mes no valido"

#el mes en número
mes = int(input("Ingrese el mes: "))
numero_cliente = int(input("Ingrese el número de cliente: "))


#-------

lista_datos = datos.to_dict(orient='records')
#------
lista_datos2 = datos2.to_dict(orient='records')
#------
lista_datos3 = datos3.to_dict(orient='records')
#------
#------

#BIEN
#CUENTA------- 2)
def Cuenta_cliente(numero_cliente):
  cu_cli = datos2[datos2['NUM_CLIENTE'] == numero_cliente]['CUENTA'].unique()
  return cu_cli

cuenta_cliente = Cuenta_cliente(numero_cliente)

#-------
#BIEN
#PRODUCTO ------- 3)

def Tarjetas_d_o_c(cuenta_cliente):
  debido_o_credito = []
  for kk in cuenta_cliente:
    for zz in range(len(lista_datos2)):
      if kk == lista_datos2[zz]['CUENTA']:
        if lista_datos2[zz]['PRODUCTO'] == 'TDC':
          nombre = 'TDC'
        else:
          nombre = 'TDD'
    debido_o_credito.append(nombre)
    A12 = debido_o_credito
  return A12


def Resulta(Lista):
  resultado = [100 if x == 'TDC' else 200 for x in Lista]
  return resultado

#------------

#BIEN
#EDAD CLIENTE------ 4)

def R_F_C(numero_cliente):
  Lista = [10,20,30,40]
  for ii in range(len(Lista)):
    if numero_cliente == Lista[ii]:
      rfc = lista_datos[ii]['RFC']

  Fe = rfc[4:10]
  año = int(Fe[0:2])
  Mes = int(Fe[2:4])
  dia = int(Fe[4:6])

  if  25 < año <= 99:
    año = año + 1900
  else:
    año = 2000

  #año,Mes,dia = R_F_C(numero_cliente)
  hoy = fecha(mes)
  fecha_nacimiento = date(año,Mes,dia)
  edad = hoy.year - fecha_nacimiento.year
  return edad


#---------------

#BIEN
#NIVEL SOCIOECONOMICO ----- 5
Dat_fes = [{10:'OCTUBRE'},{11:'NOVIEMBRE'},{12:'DICIEMBRE'}]

def NIVEL(numero_cliente):
  for ii in range(len(lista_datos)):
    for k in range(3):
      try:
        if Dat_fes[k][mes] == lista_datos[ii]['MES']:
          if lista_datos[ii]['NUM_CLIENTE'] == numero_cliente:
            xx = lista_datos[ii]['INGRESO_MENSUAL']

      except KeyError:
        pass

  if xx >= 85000:
    return 'A/B'
  elif 35000 <= xx < 85000:
    return 'C+'
  elif 11600 <= xx < 35000:
    return 'C'
  elif 6800 <= xx < 11600:
    return 'D+'
  elif 2700 <= xx < 6800:
    return 'D'
  elif 0 <= xx < 2700:
    return 'E'

#---------------------

#NUMERO DE OPERACIONES ----6)
#BIEN

def Total_operaciones(cuenta_cliente):
  Portador = []
  for ii in cuenta_cliente:
    Sumita = 0
    for ij in range(len(lista_datos3)):
      for k in range(len(Dat_fes)):
        try:
          if Dat_fes[k][mes] == lista_datos3[ij]['MES']:
            if ii == lista_datos3[ij]['CUENTA']:
               Sumita += 1
        except KeyError:
          pass
    Portador.append(Sumita)
  return Portador


#-----------

#BIEN
#SALDO ------ 7

def Saldos(mes,cuenta_cliente):
  Dat_fes = [{10:'OCTUBRE'},{11:'NOVIEMBRE'},{12:'DICIEMBRE'}]
  Saldo = []
  for ii in cuenta_cliente:
    S1 = 0
    for ij in range(len(lista_datos3)):
      for k in range(len(Dat_fes)):
        try:
          if Dat_fes[k][mes] == lista_datos3[ij]['MES']:
            if ii == lista_datos3[ij]['CUENTA']:
              if lista_datos3[ij]['TIPO OPERACIÓN'] == 1:
                S1 -= lista_datos3[ij]['IMPORTE']
              else:
                S1 += lista_datos3[ij]['IMPORTE']
        except KeyError:
          pass
    Saldo.append(S1)
  return Saldo


#----------------

# Función R (contador de meses)
#BIEN
def R(mes):
    tt = 10
    r = 0
    while tt <= mes:
        r += 1
        tt += 1
    return r

# Función para calcular el pago mínimo
def fun(mes, r, TATA,numero_cliente):
    inicio_del_mes = 0
    debito = 0  # Inicializar debito con un valor predeterminado
    for hh in range(r):
        for xx in range(len(TATA)):
            if TATA[xx] == 'TDC':
                Super_duplex = Saldos(mes - r + hh + 1, Cuenta_cliente(numero_cliente))
                if Super_duplex[xx] <= 0:
                    pago_minimo = inicio_del_mes * 0.4 + Super_duplex[xx]
                else:
                    pago_minimo = 0
            else:
                debito = 0  # Actualizar debito si es una tarjeta de débito
                pago_minimo = 0
            Pago = np.abs(pago_minimo)
        inicio_del_mes = pago_minimo

    if len(Cuenta_cliente(numero_cliente)) == 2:
        return [Pago, debito]  # Retorna una lista
    else:
        if TATA[0] == 'TDC':
            return [Pago]  # Retorna una lista con un solo elemento
        else:
            return [debito]  # Retorna una lista con un solo elemento


#--------------------

#BIEN
STD = []
Za = Tarjetas_d_o_c(Cuenta_cliente(numero_cliente))
for pi in range(len(Za)):
  STD.append({(Cuenta_cliente(numero_cliente))[pi]:(Tarjetas_d_o_c(Cuenta_cliente(numero_cliente)))[pi]})

# Función YU
def YU(cuenta_cliente):
    conta = 0
    for ii in cuenta_cliente:
        for ij in range(len(lista_datos3)):
            for k in range(len(Dat_fes)):
                try:
                    if Dat_fes[k][mes] == lista_datos3[ij]['MES']:
                        if ii == lista_datos3[ij]['CUENTA']:
                            for kk in range(len(STD)):
                                if STD[kk][ii] == 'TDC':
                                    if lista_datos3[ij]['TIPO OPERACIÓN'] == 2:
                                        conta += lista_datos3[ij]['IMPORTE']
                                else:
                                    break
                except KeyError:
                    pass
    return conta


#BIEN
# Función Cumpliento
def Cumpliento(TT,FUN,YU):
    for y in TT:
        if y == 'TDC':
            if len(TT) == 2:
                Hh, UU = FUN
                if YU >= Hh:  # Comparar con el primer elemento de la lista
                    return 'SI', 'NA'
                else:
                    return 'NO', 'NA'
            else:
                pago_minimo = FUN
                if YU >= pago_minimo[0]:  # Comparar con el primer elemento de la lista
                    return 'SI'
                else:
                    return 'NO'
        else:
            return 'NA'


# Función para obtener el número total de cuentas de todos los clientes
def total_cuentas(clientes):
    return sum(len(Cuenta_cliente(cliente)) for cliente in clientes)

# Inicialización de la matriz
Numero_de_cliente = int(input('Digite cuantos clientes son: '))
Lista_final = []
for p in range(Numero_de_cliente):
    Su_numero = int(input('Digite número de cliente: '))
    Lista_final.append(Su_numero)

mes = int(input('Digite mes: '))

# Calcular el número total de filas necesarias
num_filas = total_cuentas(Lista_final)
matriz = [[None] * 9 for _ in range(num_filas)]

# Llenar la matriz
fila_actual = 0  # Contador para la fila actual de la matriz

for Gg in Lista_final:
    cuentas = Cuenta_cliente(Gg)  # Obtener las cuentas del cliente actual
    for cuenta in cuentas:
        for j in range(9):
            if j == 0:
                matriz[fila_actual][j] = str(fecha(mes))  # Fecha
            elif j == 1:
                matriz[fila_actual][j] = cuenta  # Número de cuenta
            elif j == 2:
                matriz[fila_actual][j] = Resulta(Tarjetas_d_o_c([cuenta]))[0]  # Tipo de cuenta
            elif j == 3:
                matriz[fila_actual][j] = R_F_C(Gg)  # Edad
            elif j == 4:
                matriz[fila_actual][j] = NIVEL(Gg)  # Nivel socioeconómico
            elif j == 5:
                matriz[fila_actual][j] = Total_operaciones([cuenta])[0]  # Total de operaciones
            elif j == 6:
                matriz[fila_actual][j] = Saldos(mes, [cuenta])[0]  # Saldo
            elif j == 7:
                pago_minimo = fun(mes, R(mes), Tarjetas_d_o_c([cuenta]),Gg)
                if isinstance(pago_minimo, (list, tuple)):  # Verificar si es una lista o tupla
                  matriz[fila_actual][j] = pago_minimo[0]  # Asignar el primer elemento
                else:
                  matriz[fila_actual][j] = pago_minimo  # Asignar el valor directamente
            elif j == 8:
                cumplimiento = Cumpliento(Tarjetas_d_o_c(Cuenta_cliente(Gg)),fun(mes,R(mes),Tarjetas_d_o_c(Cuenta_cliente(Gg)),Gg),YU(Cuenta_cliente(Gg)))
                if isinstance(cumplimiento, (list, tuple)):  # Verificar si es una lista o tupla
                    matriz[fila_actual][j] = cumplimiento[0]  # Asignar el primer elemento
                else:
                    matriz[fila_actual][j] = cumplimiento  # Asignar el valor directamente
        fila_actual += 1  # Mover a la siguiente fila

# Imprimir la matriz
for fila in matriz:
    print(fila)
```

```python
import pandas as pd
import numpy as np
from datetime import date

# Cargar archivos Excel
ESTATUS_CLIENTES = "ESTATUS CLIENTES.xlsx"
ESTATUS_CTAS = "ESTATUS CTAS.xlsx"
OPERACIONES = "OPERACIONES.xlsx"

# Leer los datos de los archivos Excel
datos_clientes = pd.read_excel(ESTATUS_CLIENTES)
datos_cuentas = pd.read_excel(ESTATUS_CTAS)
datos_operaciones = pd.read_excel(OPERACIONES)

# Convertir los DataFrames a listas de diccionarios para facilitar el acceso
lista_clientes = datos_clientes.to_dict(orient='records')
lista_cuentas = datos_cuentas.to_dict(orient='records')
lista_operaciones = datos_operaciones.to_dict(orient='records')

# Definir meses y sus días correspondientes
MESES = {10: 31, 11: 30, 12: 31}

# Función para obtener la fecha correspondiente al mes
def fecha(mes):
    """Retorna la fecha correspondiente al mes ingresado."""
    if mes in MESES:
        return date(2024, mes, MESES[mes])
    else:
        return "Mes no válido"

# Solicitar mes y número de cliente
mes = int(input("Ingrese el mes (10, 11, 12): "))

#Valor arbitrario al cliente
numero_cliente=10

# Función para obtener las cuentas de un cliente
def obtener_cuentas_cliente(numero_cliente):
    """Retorna las cuentas asociadas a un cliente."""
    return datos_cuentas[datos_cuentas['NUM_CLIENTE'] == numero_cliente]['CUENTA'].unique()

# Función para determinar el tipo de tarjeta (TDC o TDD)
#PRODUCTO ------- 3)

def Tarjetas_d_o_c(cuenta_cliente):
  """Retorna una lista con el tipo de tarjeta (TDC o TDD) para cada cuenta."""
  debido_o_credito = []
  for kk in cuenta_cliente:
    for zz in range(len(lista_datos2)):
      if kk == lista_datos2[zz]['CUENTA']:
        if lista_datos2[zz]['PRODUCTO'] == 'TDC':
          nombre = 'TDC'
        else:
          nombre = 'TDD'
    debido_o_credito.append(nombre)
    A12 = debido_o_credito
  return A12

"""Retorna una lista con el tipo de tarjeta (100 o 200) para cada tarjeta"""
def Resulta(Lista):
  resultado = [100 if x == 'TDC' else 200 for x in Lista]
  return resultado



# Función para calcular la edad del cliente
def calcular_edad(numero_cliente):
    """Calcula la edad del cliente basado en su RFC."""
    for cliente in lista_clientes:
        if cliente['NUM_CLIENTE'] == numero_cliente:
            rfc = cliente['RFC']
            año = int(rfc[4:6])
            mes_nacimiento = int(rfc[6:8])
            dia_nacimiento = int(rfc[8:10])
            año = 1900 + año if año > 25 else 2000 + año
            edad = fecha(mes).year - date(año, mes_nacimiento, dia_nacimiento).year
            return edad
    return None

# Función para determinar el nivel socioeconómico
def nivel_socioeconomico(numero_cliente):
    """Retorna el nivel socioeconómico basado en el ingreso mensual."""
    for cliente in lista_clientes:
        if cliente['NUM_CLIENTE'] == numero_cliente:
            ingreso = cliente['INGRESO_MENSUAL']
            if ingreso >= 85000:
                return 'A/B'
            elif 35000 <= ingreso < 85000:
                return 'C+'
            elif 11600 <= ingreso < 35000:
                return 'C'
            elif 6800 <= ingreso < 11600:
                return 'D+'
            elif 2700 <= ingreso < 6800:
                return 'D'
            elif 0 <= ingreso < 2700:
                return 'E'
    return None



# Función para contar el total de operaciones por cuenta
def total_operaciones(cuenta_cliente):
  """Retorna el número total de operaciones por cuenta."""
  Portador = []
  for ii in cuenta_cliente:
    Sumita = 0
    for ij in range(len(lista_datos3)):
      for k in range(len(Dat_fes)):
        try:
          if Dat_fes[k][mes] == lista_datos3[ij]['MES']:
            if ii == lista_datos3[ij]['CUENTA']:
               Sumita += 1
        except KeyError:
          pass
    Portador.append(Sumita)
  return Portador


# Función para calcular el saldo de cada cuenta
def Saldos(mes,cuenta_cliente):
  """Retorna el saldo de cada cuenta."""
  Saldo = []
  for ii in cuenta_cliente:
    S1 = 0
    for ij in range(len(lista_datos3)):
      for k in range(len(Dat_fes)):
        try:
          if Dat_fes[k][mes] == lista_datos3[ij]['MES']:
            if ii == lista_datos3[ij]['CUENTA']:
              if lista_datos3[ij]['TIPO OPERACIÓN'] == 1:
                S1 -= lista_datos3[ij]['IMPORTE']
              else:
                S1 += lista_datos3[ij]['IMPORTE']
        except KeyError:
          pass
    Saldo.append(round(S1,2))
  return Saldo


# Función para calcular el pago mínimo

# Función R (contador de meses)
#BIEN
def R(mes):
    tt = 10
    r = 0
    while tt <= mes:
        r += 1
        tt += 1
    return r

# Función para calcular el pago mínimo
def calcular_pago_minimo(mes, r, TATA,numero_cliente):
    inicio_del_mes = 0
    debito = 0  # Inicializar debito con un valor predeterminado
    for hh in range(r):
        for xx in range(len(TATA)):
            if TATA[xx] == 'TDC':
                Super_duplex = Saldos(mes - r + hh + 1, Cuenta_cliente(numero_cliente))
                if Super_duplex[xx] <= 0:
                    pago_minimo = inicio_del_mes * 0.4 + Super_duplex[xx]
                else:
                    pago_minimo = 0
            else:
                debito = 0  # Actualizar debito si es una tarjeta de débito
                pago_minimo = 0
            Pago = round(np.abs(pago_minimo),2)
        inicio_del_mes = pago_minimo

    if len(Cuenta_cliente(numero_cliente)) == 2:
        return [Pago, round(debito,2)]  # Retorna una lista
    else:
        if TATA[0] == 'TDC':
            return [Pago]  # Retorna una lista con un solo elemento
        else:
            return [round(debito,2)]  # Retorna una lista con un solo elemento

# Función para verificar el cumplimiento de pagos
def verificar_cumplimiento(cuentas, tipos_tarjeta, pagos_minimos,mes):
    """Verifica si se cumplió con el pago mínimo."""
    cumplimiento = []
    for cuenta, tipo, pago in zip(cuentas, tipos_tarjeta, pagos_minimos):
        if tipo == 'TDC':
            saldo = Saldos(mes,[cuenta])[0]
            cumplimiento.append('SI' if saldo >= pago else 'NO')
        else:
            cumplimiento.append('NA')  # No aplica para débito
    return cumplimiento

# Función principal para generar la matriz de resultados
def generar_matriz(clientes, mes):
    """Genera una matriz con los datos de los clientes y sus cuentas."""
    matriz = []
    for cliente in clientes:
        cuentas = obtener_cuentas_cliente(cliente)
        tip = Tarjetas_d_o_c(cuentas)
        tipos = Resulta(tip)
        edades = [calcular_edad(cliente)] * len(cuentas)
        niveles = [nivel_socioeconomico(cliente)] * len(cuentas)
        operaciones = total_operaciones(cuentas)
        saldos = Saldos(mes,cuentas)
        pagos_minimos = calcular_pago_minimo(mes, R(mes), Tarjetas_d_o_c(cuentas),cliente)
        cumplimiento = verificar_cumplimiento(cuentas, tip, pagos_minimos,mes)

        for i in range(len(cuentas)):
            fila = [
                str(fecha(mes)),  # Fecha
                cuentas[i],  # Número de cuenta
                tipos[i],  # Tipo de tarjeta
                edades[i],  # Edad
                niveles[i],  # Nivel socioeconómico
                operaciones[i],  # Total de operaciones
                saldos[i],  # Saldo
                pagos_minimos[i],  # Pago mínimo
                cumplimiento[i]  # Cumplimiento
            ]
            matriz.append(fila)
    return matriz

# Solicitar número de clientes y sus números
Numero_de_cliente = int(input('Digite cuántos clientes son: '))
Lista_final = [int(input('Digite número de cliente: ')) for _ in range(Numero_de_cliente)]

# Generar y mostrar la matriz
matriz = generar_matriz(Lista_final, mes)
for fila in matriz:
    print(fila)
```

```python
# Función para guardar la matriz en un archivo de texto plano con datos centrados
def guardar_matriz_en_txt(matriz, nombre_archivo):
    """Guarda la matriz en un archivo de texto plano, centrando los valores en cada columna."""
    # Definir el ancho de cada columna
    anchos = [12, 10, 8, 6, 6, 8, 12, 12, 8]  # Ajusta estos valores según tus necesidades

    with open(nombre_archivo, 'w') as archivo:
        for fila in matriz:
            # Formatear cada valor para que esté centrado en su columna
            fila_formateada = []
            for i, valor in enumerate(fila):
                valor_str = str(valor)
                espacio_extra = anchos[i] - len(valor_str)
                izquierda = espacio_extra // 2
                derecha = espacio_extra - izquierda
                valor_centrado = ' ' * izquierda + valor_str + ' ' * derecha
                fila_formateada.append(valor_centrado)

            # Unir las columnas con '|' y escribir la fila en el archivo
            fila_str = '|'.join(fila_formateada)
            archivo.write(fila_str + '\n')

# Solicitar número de clientes y sus números
Numero_de_cliente = int(input('Digite cuántos clientes son: '))
Lista_final = [int(input('Digite número de cliente: ')) for _ in range(Numero_de_cliente)]

# Generar la matriz
matriz = generar_matriz(Lista_final, mes)

# Guardar la matriz en un archivo de texto plano
nombre_archivo = 'matriz_clientes.txt'
guardar_matriz_en_txt(matriz, nombre_archivo)

print(f"La matriz se ha guardado en el archivo: {nombre_archivo}")
```
